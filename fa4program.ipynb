```python
#password checker

#Imports
from IPython.display import clear_output
import time

#Main
while True:
  print("Please input password (Hint, the password is very hard): ")
  password = input("Password: ")
  if password == "veryhard":
    print("Nice! You got it correct!")
    break
  else:
    print("Wrong! Try again!")
    time.sleep(2)
    clear_output()
    continue
```
